```python
# PROJECT 3 - AI RECOMMENDATION LOGIC

print("Welcome to Movie Recommendation System")

while True:
    choice = input("\nEnter your favorite genre (action, comedy, horror) or type 'exit' to quit: ").lower()
    if choice == "action":
        print("Recommended Movies:")
        print("Avengers")
        print("John Wick")
    elif choice == "comedy":
        print("Recommended Movies:")
        print("Mr. Bean")
        print("Home Alone")
    elif choice == "horror":
        print("Recommended Movies:")
        print("The Conjuring")
        print("Insidious")
    elif choice == "exit":
        print("Thanks for using the system!")
        break
    else:
        print("Sorry, no recommendations found.")
```
